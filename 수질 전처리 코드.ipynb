```python
import pandas as pd
import numpy as np
```

## 1. Total code

```python
지역리슽=["강원","경기","경남","경북","광주","대구","대전","부산","서울",
     "세종","울산","인천","전남","전북","제주","충남","충북"]
```

```python
for io in range(len(지역리슽)):
    액셀이름 = 지역리슽[io]
    data = pd.read_excel("./19년/{}.xls".format(액셀이름))
    data.drop(['페놀(㎎/L)','카드뮴(㎎/L)', '시안(㎎/L)', '납(㎎/L)',
           '6가크롬(㎎/L)', '비소(㎎/L)', '수은(㎎/L)', '구리(㎎/L)', '아연(㎎/L)', '크롬(㎎/L)',
           '니켈(㎎/L)', '바륨(㎎/L)', '셀레늄(㎎/L)','헥사클로로벤젠 (㎍/L)','불소(㎎/L)', '색도(㎎/L)', '노말헥산추출물질(㎎/L)', '용해성망간(㎎/L)',
           '용해성철(㎎/L)', '음이온계면활성제(㎎/L)', '트리클로로에틸렌(㎎/L)', '테트라클로로에틸렌(㎎/L)',
           '사염화탄소', '1.2-디클로로에탄(㎎/L)', '디클로로메탄(㎎/L)', '벤젠(㎎/L)',
           '폴리크로리네이티트비페닐(㎎/L)', '유기인(㎎/L)', '안티몬 (㎎/L)', '클로로포름 (㎎/L)',
           '디에틸헥실프탈레이트 (㎎/L)', '1.4-다이옥세인 (㎎/L)', '투명도(m)','유량(㎥/s)', '포름알데히드(㎎/L)'
              ],1,inplace=True)
    data.reset_index(drop=True)
    data2 = data.groupby("년/월").mean()
    data2=pd.DataFrame(np.array(data2),columns=['수소이온농도', '용존산소', 'BOD', 'COD', '부유물질', '총질소', '총인', 'TOC', '수온',
           '전기전도도', '총대장균군수', '용존총질소', '암모니아성 질소', '질산성 질소', '용존총인', '인산염인',
           '클로로필', '분원성대장균군수'])
    data2.reset_index(drop=True).to_excel("./19년/{}전처리.xlsx".format(액셀이름))
```

```python
컬러엄 = ['년/월', '수소이온농도', '용존산소', 'BOD', 'COD', '부유물질', '총질소', '총인', 'TOC', '수온',
       '전기전도도', '총대장균군수', '용존총질소', '암모니아성 질소', '질산성 질소', '용존총인', '인산염인',
       '클로로필', '분원성대장균군수']
list_날짜 = []
for i in range(len(data)):
    list_날짜.append(data.iloc[i,0])
list_전철 = list(set(list_날짜))
list_전철.sort()
def get(날짜,칼럼):
    sum_,cnt = 0,0
    for i in range(len(data)):
        if data.iloc[i,0] == list_전철[날짜]:
            sum_ += data.iloc[i,칼럼]
            cnt+=1
    return sum_ if sum_ == 0 else round(sum_/cnt,2)
list_avg = []
for i in range(len(list_전철)):
    list_avg.append([list_전철[i],get(i,1),get(i,2),get(i,3),get(i,4),get(i,5),get(i,6),get(i,7),
                     get(i,8),get(i,9),get(i,10),get(i,11),get(i,12),get(i,13),get(i,14),get(i,15),get(i,16),get(i,17),get(i,18)])
df = pd.DataFrame(list_avg,columns=컬러엄)
df.reset_index(drop=True).to_excel("./전처리/{}전처리.xlsx".format(액셀이름))
```
