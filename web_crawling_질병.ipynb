```python
from selenium import webdriver as wb
import pandas as pd
from bs4 import BeautifulSoup as bs
import requests as req
import time
```

```python
driver = wb.Chrome()
driver.get('http://www.cdc.go.kr/npt/biz/npp/ist/bass/bassAreaStatsMain.do')
```

```python
def datetime(start,end) :
    driver.find_element_by_css_selector('input#areaDissFrmStartDt').click()
    driver.find_element_by_css_selector('button.ui-datepicker-close').click()
    driver.find_element_by_css_selector('input#areaDissFrmStartDt').send_keys(start)
    driver.find_element_by_css_selector('input#areaDissFrmEndDt').click()
    driver.find_element_by_css_selector('button.ui-datepicker-close').click()
    driver.find_element_by_css_selector('input#areaDissFrmEndDt').send_keys(end)
def check_box():
    driver.get('http://www.cdc.go.kr/npt/biz/npp/ist/bass/bassAreaStatsMain.do')
    driver.find_element_by_css_selector('#areaDissFrm_searchType2').click()
    time.sleep(0.5)
    driver.find_element_by_css_selector('div.ms-parent.multi.w65 > button').click()
    driver.find_element_by_css_selector('div.ms-parent.multi.w65 > div > ul > li:nth-child(2) > label').click()
    driver.find_element_by_css_selector('#areaDissFrm > div > ul.form-addr.mw30 > li.select-cnt.mfr.mbw80 > div:nth-child(3) > button > span').click()
    driver.find_element_by_css_selector('#areaDissFrm > div > ul.form-addr.mw30 > li.select-cnt.mfr.mbw80 > div:nth-child(3) > div > ul > li.ms-select-all > label').click()
    driver.find_element_by_css_selector('#areaDissFrm > div > ul.form-addr.mw30 > li.select-cnt.mfr.mbw80 > div:nth-child(7) > button > div').click()
    time.sleep(1)
    driver.find_element_by_css_selector('#areaDissFrm > div > ul.form-addr.mw30 > li.select-cnt.mfr.mbw80 > div:nth-child(7) > div > ul > li.ms-select-all > label').click()
    driver.find_element_by_css_selector('#areaDissFrm_areaType2').click()

def search():
    driver.find_element_by_css_selector('#areaDissFrm > input.btn-select.bassAreaStatsBtn').click()
```

```python
money = {
     '날짜': [], '큰지역' : [], '작은지역' : [], '콜레라' : [], 
     '장티푸스' : [], '파라티푸스' : [], '세균성이질' : [], 
     '장출혈성대장균감염증' : [], 'A형간염' : []
}
```

```python
def get_data(money):
    start = '2010-01-01'
    end = '2010-02-01'
    check_box()
    while True:
        datetime(start,end)
        search()
        soup = bs(driver.page_source,'html.parser')
        i = 1
        time.sleep(1.5)
        while True:
            i += 1
            try :
                soup.select_one('#table > tbody > tr:nth-child({}) > td:nth-child(1)'.format(i)).text
                print('가능')
            except :
                print('불가능')
                break
            money['날짜'].append(start[:-3])
            money['큰지역'].append(soup.select_one('#table > tbody > tr:nth-child({}) > td:nth-child(1)'.format(i)).text)
            money['작은지역'].append(soup.select_one('#table > tbody > tr:nth-child({}) > td:nth-child(2)'.format(i)).text)
            money['콜레라'].append(soup.select_one('#table > tbody > tr:nth-child({}) > td:nth-child(3)'.format(i)).text)
            money['장티푸스'].append(soup.select_one('#table > tbody > tr:nth-child({}) > td:nth-child(4)'.format(i)).text)
            money['파라티푸스'].append(soup.select_one('#table > tbody > tr:nth-child({}) > td:nth-child(5)'.format(i)).text)
            money['세균성이질'].append(soup.select_one('#table > tbody > tr:nth-child({}) > td:nth-child(6)'.format(i)).text)
            money['장출혈성대장균감염증'].append(soup.select_one('#table > tbody > tr:nth-child({}) > td:nth-child(7)'.format(i)).text)
            money['A형간염'].append(soup.select_one('#table > tbody > tr:nth-child({}) > td:nth-child(8)'.format(i)).text)
        start = end
        if int(end.split('-')[1])+1 == 13:
            end = str(int(end.split('-')[0])+1) + '-01-01'
        else:
            if len(str(int(end.split('-')[1])+1)) == 2:
                end = end[:5] + str(int(end.split('-')[1])+1) + '-01'
            else :
                end = end[:5] +'0' + str(int(end.split('-')[1])+1) + '-01'
        if end == '2019-03-01':
            return money
    
```

```python
money = get_data(money)
```
