```python
data=pd.read_excel("./액셀/서울_구단위_스케일링.xlsx")
지역리=list(data["지역"].values)
지역릿=list(set(지역리))
지역릿.sort()
```

```python
plt.figure(figsize=(20,8))
for cnt in tqdm_notebook(range(len(지역리))):
    plt.plot(data[data["지역"]==지역리[cnt]]["환자수"],lw=0.01,marker="o",ls="-.",markersize=8)
plt.legend(지역릿)
plt.title("서울 구단위 지역별 A형간염 발생자 수[2011년~2018년]",pad=15,fontsize=25)
plt.xlabel("지역",fontsize="xx-large")
plt.ylabel("발생자 수 \n [scaler]",rotation=-360,labelpad=30,fontsize="xx-large")
plt.xticks(ticks=[0,100,200,300,400,500,600,700],labels=["지역A","지역B","지역C","지역D","지역E","지역F","지역G","지역H"],fontsize=15)
plt.grid()


#plt.savefig("저장.png")
```
